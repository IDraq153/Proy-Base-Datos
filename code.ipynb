## Generación de Inserts
### Entidades fuertes

### Providers
Listar los proveedores de los pacientes sin repeticiones

```python
tmp = patients['insurance_provider']
providers = list(tmp.unique())
providers
```

```python
text = '-- migrate:up \n\n'
id = 1
for n in providers:
  text = text + f"INSERT INTO providers (id, name) VALUES ({id}, '{n}');\n"
  id = id + 1
text = text + '\n-- migrate:down \n\nDELETE FROM providers;'
with open('inserts_providers.sql', 'w') as archivo:
  archivo.write(text)
```
